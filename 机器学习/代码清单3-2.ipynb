```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x-2)*x*(x+2)
```

```python
print(f(1))
```

```python
print(f(np.array([1,2,3])))
```

```python
x=np.arange(-3,3.5,0.5)
print(x)

x=np.arange(-3,3,0.5)
print(x)
```

```python
x=np.linspace(-3,3,10)
print(np.round(x,2))
```

```python
plt.plot(x,f(x))
plt.show()
```

```python
# 定义函数
def f2(x,W):
    return (x-W)*x*(x+2)

# 定义X
x = np.linspace(-3,3,100) # 把x分为100份

# 绘制图形
plt.plot(x,f2(x,2),color='black',label='w=2')
plt.plot(x,f2(x,1),color='cornflowerblue',label='w=1')
plt.legend(loc="upper left") # 显示图例
plt.ylim(-15,15) # y轴范围
plt.title('f_2(x)') # 标题
plt.xlabel('x') # x标签
plt.ylabel('y') # y标签
plt.grid(True) # 网格线
plt.show()
```
