```python
#2
#高斯函数
def gauss (x,mu,s):
    return np.exp(-(x-mu)**2/(2*s**2))
```

```python
#主处理
M = 4
plt.figure(figsize=(4,4))
mu = np.linspace(5,30,M)
s = mu[1] - mu[0]
xb = np.linspace(X_min,X_max,100)

for j in range(M):
    y = gauss(xb,mu[j],s)
    plt.plot(xb,y,color='gray',linewidth=3)

plt.grid(True)
plt.xlim(X_min,X_max)
plt.ylim(0,1.2)
plt.show()
```
