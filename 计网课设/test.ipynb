```python
from scapy.all import *


def packet_handler(packet):
    # 打印整个数据包对象
    print(packet.show())

    # 获取和打印以太网层信息
    ether_layer = packet.getlayer(Ether)
    if ether_layer:
        print("Source MAC:", ether_layer.src)
        print("Destination MAC:", ether_layer.dst)

    # 获取和打印 IP 层信息
    ip_layer = packet.getlayer(IP)
    if ip_layer:
        print("Source IP:", ip_layer.src)
        print("Destination IP:", ip_layer.dst)

    # 获取和打印 TCP 层信息
    tcp_layer = packet.getlayer(TCP)
    if tcp_layer:
        print("Source Port:", tcp_layer.sport)
        print("Destination Port:", tcp_layer.dport)


# 使用过滤器捕获数据包，回调函数为 packet_handler
sniff(filter="tcp", prn=packet_handler, count=1)
```

```python
from scapy.all import *
from scapy.layers.l2 import Ether

captNum = 0

def packet_callback(packet):
    global captNum
    captNum = captNum + 1
    print("捕获到第%d个以太帧" %captNum)
    print("目的MACt地址为：%s" %packet[Ether].dst)
    print("源MAC地址为：%s" %packet[Ether].src)
    print("协议类型为：%s" %packet[Ether].type)
    print("payload:\n%s" %hexdump(packet[Ether].payload,dump = True))

print(sniff(prn = packet_callback,count = 5))
```
